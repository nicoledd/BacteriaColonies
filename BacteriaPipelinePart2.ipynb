**INPUT:** The raw NEFI output, consisting of a folder for each bacteria image

**OUTPUT:** Extracts the network text file from each folder

```python
# GLOBAL VARIABLES AND IMPORTED MODULES

# used to read names of all raw bacteria images
#    and to create new directories 
import os

nefi_directory_name = 'Step 3 - nefi output/'
text_directory_name = 'Step 4 - NEFI Text Files/'
```

```python
# create a new directory
try:
    os.mkdir(text_directory_name)
except:
    pass

with os.scandir(nefi_directory_name) as folders:
    for folder in folders:
        with os.scandir(folder) as outputs:
            for output in outputs:
                if(output.name[-3:] == 'txt'):
                    os.rename(nefi_directory_name + folder.name + '/' + output.name, text_directory_name + output.name)
                    print(output.name)
```
